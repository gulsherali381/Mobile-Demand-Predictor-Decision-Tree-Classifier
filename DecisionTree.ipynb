```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
df_phone=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/AI Enginer/mobile_train_70.csv')
df_phone.head(15)
```

```python
model=DecisionTreeClassifier()
features=['RAM_GB','Price_PKR','Camera_MP','Battery_mAh','Storage_GB']
output=['Demand']
model.fit(df_phone[features],df_phone[output])
```

```python
#Model Prediction
pred=model.predict([[8,75000,50,5000,128]])
for pred in pred:
    if pred==1:
        print("High Demand")
    else:
        print("Low Demand")
print(f"Model Prediction:{pred}")
print(f"Model Accuracy:{model.score(df_phone[features],df_phone[output])}")
```
